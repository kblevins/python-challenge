# Weather and the Equator

## Analysis
* **Trend 1**
    The hottest temperatures are, indeed, centered around the equator. They are hottest from about -25 to 25 degrees of latitude.
* **Trend 2**
    Earth is top-heavy - cities extend further to the north than they do to the south (very little weather data below -50 degrees of latitude as compared to above 50 degrees - probably related to the much larger amount of landmass in the northern hemisphere :).
* **Trend 3**
    Humidity, cloudiness, and wind speed appear to be unrelated to latitude.

```python
# import dependencies
import pandas as pd
import json
from matplotlib import pyplot as plt
from citipy import citipy
import numpy as np
import time
import random
import requests
from config import api_key
```

## Gather current weather data for 500 random cities

```python
# Randomly select at least 500 unique (non-repeat) cities based on latitude and longitude.

# create lists to store city info
city = []
country = []

# create list to store cities without weather data
no_weather = []

# store query parameters
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# create list to store weather json objects
weather_data = []

# initialize counter
city_count = 0

# loop through lat/long combos, look up current weather for the nearest city
# & add the json to weather_data until there are 500
while city_count <501:
    
    # generate random lat & long
    lat_r = random.randint(-90,90)
    lng_r = random.randint(-180,180)
    
    # find nearest city
    city_data = citipy.nearest_city(lat_r, lng_r) 
    
    # if the city name hasn't already been added to city or no_weather, send a request to 
    # openweather to retrieve the current weather data
    if (city_data.city_name not in city) and (city_data.city_name not in no_weather):
        
        # store city name & country code to use in query
        city_country = f'{city_data.city_name},{city_data.country_code}'
        
        # build query
        query_url = url + "appid=" + api_key + "&units=" + units + "&q=" + city_country
        
        # retrieve weather
        response = requests.get(query_url).json()
        
        # sleep for 1.5 seconds to avoid exceeding query limits
        time.sleep(1.5)
        
        # if there is no weather for the city, add the city name to no_weather
        if response.get('message'):
            no_weather.append(city_data.city_name)
        # otherwise, add city name to city, country code to country & weather json to weather_data
        # also print the count, the city name & add one to the counter
        else: 
            city.append(city_data.city_name)
            country.append(city_data.country_code)
            weather_data.append(response)
            print(f'City {city_count} is {city_data.city_name}, {city_data.country_code}')
            city_count+=1
    
```

```python
# create dataframe with weather data for each city
city_df = pd.DataFrame(columns = ['City'], index = np.arange(500))

for x in np.arange(500):
    city_df.set_value(x, 'City', weather_data[x]['name'])
    city_df.set_value(x, 'Cloudiness', weather_data[x]['clouds']['all'])
    city_df.set_value(x, 'Country', weather_data[x]['sys']['country'])
    city_df.set_value(x, 'Date', weather_data[x]['dt'])
    city_df.set_value(x, 'Humidity', weather_data[x]['main']['humidity'])
    city_df.set_value(x, 'Lat', weather_data[x]['coord']['lat'])
    city_df.set_value(x, 'Lng', weather_data[x]['coord']['lon'])
    city_df.set_value(x, 'Max Temp', weather_data[x]['main']['temp_max'])
    city_df.set_value(x, 'Wind Speed', weather_data[x]['wind']['speed'])
    
city_df.head()
```

```python
# write df to csv
city_df.to_csv('city data.csv')
```

## Latitude vs Temperature Plot

```python
plt.axes(facecolor='lightgray')
plt.scatter(city_df['Lat'], city_df['Max Temp'], marker = '.', color = 'magenta', 
            edgecolor='black', linewidth= .7, zorder = 2)
plt.grid(b = True, which='major', color='w', zorder = 1)
plt.xlim(-90, 100)
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.title("City Latitude vs Max Temperature\nNovember 16, 2017")
plt.savefig("lat_v_temp.png")
plt.show()
```
