```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

folder = 'C:/Users/Kali/repo/python-challenge'

csv_file1 = folder + '/Pymaceuticals/raw_data/clinicaltrial_data.csv'

csv_file2 = folder + '/Pymaceuticals/raw_data/mouse_drug_data.csv'

trial = pd.read_csv(csv_file1)

drug = pd.read_csv(csv_file2)

mice = pd.merge(trial, drug)

print(mice.shape)
mice.head()
```

```python
# mouse g989 is on 2 drugs - Stelasyn & Propriva. To maintain a 1:1 mouse to drug ratio,
# this mouse should be removed from the dataset
mice_clean = mice.loc[mice['Mouse ID'] != 'g989',:]
mice_clean.shape
```

```python
# create list of drugs of interest
drug_list = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

# create subset of data with drugs of interest
mice_s = mice_clean[mice_clean.Drug.isin(drug_list)]
```

## Tumor Response to Treatment

```python
# create dataframe with the mean of tumor volume by drug & timepoint
tumor_mean = pd.DataFrame(mice_s.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean())
tumor_mean.head()
```

```python
# create dataframe with the standard error of tumor volume by drug & timepoint
tumor_sem = pd.DataFrame(mice_s.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem())
tumor_sem.head()
```

```python
# create modified dataframe with columns for each drug populated with mean tumor volume by timepoint
tumor_mean_p = tumor_mean.reset_index()
tumor_mean_p = tumor_mean_p.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')
tumor_mean_p
```

```python
# create modified dataframe with columns for each drug populated by standard error by timepoint
tumor_sem_p = tumor_sem.reset_index()
tumor_sem_p = tumor_sem_p.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# create list of x values to use from the index of tumor_mean_p df
x_values = [int(timept) for timept in tumor_mean_p.index]

# create a scatterplot with error bars for the mean tumor volume of the mice treated 
#with each drug of interest
for drug in drug_list:
    plt.errorbar(x_values, tumor_mean_p[drug], yerr=tumor_sem_p[drug],
            marker = "o", linestyle = '--')

# format plot
plt.xticks(x_values)
plt.legend()
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Volume in Response to Treatment")
plt.grid(b=True, which='major', color='b', linestyle=':')
plt.show()
```

## Metastatic Response to Treatment

```python
# create dataframe with the mean of the number of metastatic sites by drug & timepoint
met_mean = pd.DataFrame(mice_s.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean())
met_mean.head()
```

```python
# create dataframe with the standard error of the number of metastatic sites by drug & timepoint
met_sem = pd.DataFrame(mice_s.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem())
met_sem.head()
```

```python
# create modified dataframe with columns for each drug populated with mean number of metastatic sites by timepoint
met_mean_p = met_mean.reset_index()
met_mean_p = met_mean_p.pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')
met_mean_p
```

```python
# create modified dataframe with columns for each drug populated with standard error for the 
# number of metastatic sites by timepoint
met_sem_p = met_sem.reset_index()
met_sem_p = met_sem_p.pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')

# create a scatterplot with error bars for the mean number of metastatic sites of the mice treated 
#with each drug of interest
for drug in drug_list:
    plt.errorbar(x_values, met_mean_p[drug], yerr=met_sem_p[drug],
            marker = "o", linestyle = '--')

plt.xticks(x_values)
plt.legend()
plt.xlabel("Time (Days)")
plt.ylabel("Metastatic Sites")
plt.title("Metastatic Spread in Response to Treatment")
plt.grid(b=True, which='major', color='b', linestyle=':')
plt.show()
```

## Survival Rates

```python
# create dataframe with the count of mice by drug & timepoint (dataframe of surviving mice at each timepoint)
survivors = pd.DataFrame(mice_s.groupby(['Drug','Timepoint'])['Mouse ID'].count())
survivors.columns = ['Mouse Count']
survivors.head()
```

```python
# create modified dataframe with columns for each drug populated with the number of surviving
# mice by timepoint
survivors_p = survivors.reset_index()
survivors_p = survivors_p.pivot(index = 'Timepoint', columns = 'Drug', values = 'Mouse Count')
survivors_p
```

```python
# calculate the percentage of surviving mice for each drug & timepoint
survivor_pct = survivors_p.applymap(lambda x: (x/25)*100)

# create lists of markers & colors to use in scatterplot for each drug
markers = ["o", "^", "s", "d"]
colors = ['red', 'blue', 'green', 'black']

# create scatterplot for the percent of surviving mice for each drug
for drug in drug_list:
    plt.plot(x_values, survivor_pct[drug],
                marker = markers[drug_list.index(drug)], 
                color = colors[drug_list.index(drug)], 
                linestyle = '--')

plt.xticks(x_values)
plt.legend()
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title("Survival Rate During Treatment")
plt.grid(b=True, which='major', color='b', linestyle=':')
plt.show()
```

```python
# create dataframe to use for barchart
bar_df = tumor_mean_p.loc[[0, 45],]

# calculate the percent change tumor size from the beginning to the 
# end of the study period for each drug
bar_df_pct = pd.DataFrame(((bar_df.loc[45]-bar_df.loc[0])/bar_df.loc[0])*100)
bar_df_pct.columns = ['Pct Change']
bar_df_pct
    
    
```

```python
# create x & y arrays for bar plot of the percent tumor change from the beginning
# to the end of the study period
x = np.arange(0,4)
y = bar_df_pct['Pct Change']

# create arrays to use in subsetting the positive & negative changes
mask1 = y < 0
mask2 = y >= 0

# plot each subset (positive & negative changes) with different formatting
plt.bar(x[mask1], y[mask1], color = 'green', width = 1, edgecolor='black', linewidth=2)
plt.bar(x[mask2], y[mask2], color = 'red', width = 1, edgecolor='black', linewidth=2)

# format plot
plt.xticks(x, bar_df_pct.index)
plt.ylabel("% Tumor Change")
plt.title("Tumor Change Over 45 Day Treatment")
plt.grid(b=True, which='major', color='black', linestyle=':')

for i, v in enumerate(y):
    if v < 0:
        plt.text(i-0.15, -5, '{}%'.format(int(v)), color='white', fontweight='bold')
    else:
        plt.text(i-0.1, 5, '{}%'.format(int(v)), color='white', fontweight='bold')    
    
plt.show()
```
